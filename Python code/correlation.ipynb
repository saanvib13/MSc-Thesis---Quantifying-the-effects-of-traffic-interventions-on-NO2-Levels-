```python
import pandas as pd
import matplotlib.pyplot as plt
from zoneinfo import ZoneInfo
import seaborn as sns
```

```python
df=pd.read_csv('../mydata/UCC_NO2(in).csv')
df.head()
```

```python
df.drop(columns=['ind','ind.1','ind.2','ind.3','ind.4'], axis=1, inplace=True)
df
```

```python
# how to rename a column in a dataframe in python?
```

```python
df.rename(columns={'date':'Timestamp'}, inplace=True)
```

```python
df
```

```python
# Convert 'Timestamp' to datetime in UTC
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%Y %H:%M', errors='coerce', utc=True)

# Drop rows with invalid timestamps or missing NO2 values
df = df.dropna(subset=['Timestamp', 'NO2'])

# Convert to Irish local time (handles DST automatically)
df['Timestamp'] = df['Timestamp'].dt.tz_convert(ZoneInfo("Europe/Dublin"))

# Sort by timestamp
df.sort_values('Timestamp', inplace=True)

# Set timestamp as index
# df.set_index('Timestamp', inplace=True)
df
```

```python
df.reset_index(inplace=True)
df['date']=pd.to_datetime(df['Timestamp']).dt.date
df['hour']=pd.to_datetime(df['Timestamp']).dt.hour

df_22_24=df[((df['date']>=pd.to_datetime('2022-01-01').date()) & (df['date']<=pd.to_datetime('2024-12-31').date()))]
df_22_24
```

```python
df_22_24.reset_index(inplace=True)
```

```python
df_22_24.drop(columns=['index','level_0'],axis=1, inplace=True)
```

```python
df_22_24
```

```python

# Extract date-time components
df_22_24['year'] = df_22_24['Timestamp'].dt.year
df_22_24['month'] = df_22_24['Timestamp'].dt.month
df_22_24['hour'] = df_22_24['Timestamp'].dt.hour

# Define season based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df_22_24['season'] = df_22_24['month'].apply(get_season)
df_22_24
```

```python
# View missing values count
print(df_22_24.isnull().sum())

# Option 1: Drop rows with missing NO2 or meteorological data
df_clean = df_22_24.dropna(subset=['NO2', 'temp', 'rhum', 'wdsp'])

# Option 2: (Optional) Fill missing weather data with interpolation
# df['temp'] = df['temp'].interpolate(method='time')
```

```python



# Select NO2 and meteorological variables for correlation
corr_vars = ['NO2', 'temp', 'rhum', 'wdsp', 'vappr', 'rain','dewpt']  # Adjust based on your columns
corr_matrix = df_clean[corr_vars].corr()

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: NOâ‚‚ vs Meteorological Variables')
plt.tight_layout()
plt.show()
```
